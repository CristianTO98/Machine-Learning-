Como se observa, se tienen 10 etiquetas, por lo que se deben de usar modelos multiclases. El proceso que se va a seguir es el entrenamiento de diversos modelos de clasificación multiclases que se encuentran en la libreria sklearn. Para estandarizar y hacer más eficientes estos entrenamientos, se va a hacer uso de piplines y multiprocesamiento. 

Los modelos a utilizar son los siguientes:


* Árboles de decisión: se utilizan para problemas de clasificación y también para problemas de regresión. Son fáciles de interpretar y explicar, y pueden manejar características numéricas y categóricas.

* Random Forest: es un conjunto de árboles de decisión que se utilizan para problemas de clasificación y también para problemas de regresión. Pueden manejar características numéricas y categóricas, y son robustos a valores atípicos y ruido en los datos.

* Support Vector Machines (SVM): se utilizan para problemas de clasificación binaria o multiclase y también para problemas de regresión. SVM encuentra la mejor separación lineal o no lineal entre las diferentes clases.

* K vecinos más cercanos (KNN): se utiliza para problemas de clasificación y regresión. Busca los k vecinos más cercanos en el espacio de características y utiliza su etiqueta para hacer una predicción.

* Logistic Regression: La regresión logística es un modelo de aprendizaje supervisado utilizado para la clasificación binaria o multiclase. Se utiliza para predecir la probabilidad de una variable categórica dependiente en función de una o más variables independientes continuas o categóricas. El modelo utiliza una función sigmoide para convertir una entrada lineal en una salida no lineal entre 0 y 1, que representa la probabilidad de pertenecer a cada clase.

* Linear Discriminant Analysis (LDA) es un algoritmo de aprendizaje supervisado utilizado en análisis de datos y aprendizaje automático para clasificar observaciones en diferentes grupos o categorías. LDA se basa en el teorema de Bayes para estimar la probabilidad de que una observación pertenezca a una categoría determinada. 

* Quadratic Discriminant Analysis (QDA) es otro algoritmo de aprendizaje supervisado utilizado en análisis de datos y aprendizaje automático para clasificar observaciones en diferentes grupos o categorías. Al igual que LDA, QDA también se basa en el teorema de Bayes para estimar la probabilidad de que una observación pertenezca a una categoría determinada.

* MLPClassifier: red neuronal compuesta por una capa de entrada, una o más capas ocultas y una capa de salida. Cada capa contiene un número de neuronas o nodos que procesan la información recibida de la capa anterior y la propagan a la siguiente capa.

```python
from joblib import Parallel, delayed

# Se define una función para ejecutar la búsqueda de grid para una pipeline dada
def search_pipeline(pipeline, param_grid):
    grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    return grid_search

# Se crea una lista de trabajos a ejecutar
jobs = [delayed(search_pipeline)(pipeline, param_grid[name]) for name, pipeline in pipelines]

# Se ejecuta los trabajos en paralelo
results = Parallel(n_jobs=-1)(jobs)

```

#### A continuación se va a realizar la representación de la curva ROC así como del area bajo dicha curva, el ROC AUC, para los dos modelos seleccionados anteriormente. Este tipo de métricas están ideadas para clasificadores binarios, para adaptarlas a clasificadores multiclase se pueden realizar dos tipos de estrategias OvO Y OvR.

#### Con la estrategia OvO (One vs One) se representan todas las permutaciones posibles, por lo que por cada modelo se deberían de representar 90  gráficas. Por ello para este ejercicio se optará por OvR(One vs Rest) donde representaremos cada etiqueta con respecto a todas las demás, obteniendo tanto la distribución de probabilidades como la curva y área ROC para cada etiqueta. De esta manera, obtendremos 10 curvas diferentes para cada modelo, una por etiqueta.

#### En primer lugar, se debe de crear una función para calcular las coordenadas de la curva ROC, es decir, el valor de TPR y FPR. TPR es el ratio de valores positivos que realmente lo son mientras que FPR es el ratio de los valores que han salido positivos pero realmente no lo son:



```python
def calculate_tpr_fpr(y_real, y_pred):
  
    # Se calcula la matriz de confusión, y se extraen los valores de verdaderos y falsos, positivos y negativos de ella.
    cm = confusion_matrix(y_real, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    
    # Ya se tiene todo los necesario para calcular TPR Y FPR
    tpr =  TP/(TP + FN) # Sensibilidad 
    fpr = 1 - TN/(TN + FP) # Especificidad
    
    return tpr, fpr
```

#### A continuación se crea una función que calcula las coordenadas anteriores pero para cada umbral de calisificación:

```python
def get_all_roc_coordinates(y_real, y_proba):
    # Se crean las listas necesarias para guardar las diferentes coordenadas
    tpr_list = []
    fpr_list = []
    
    for i in range(len(y_proba)):
        # Se establece el umbral
        threshold = y_proba[i] 
        # Se predice en función del umbral
        y_pred = y_proba >= threshold
        # Se calcula los tpr y fpr a partir de la función anterior
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

#### Se crea la función para crear la gráfica ROC 

```python
def plot_roc_curve(tpr, fpr, scatter = True, ax = None):
 
    if ax == None:
        plt.figure(figsize = (5, 5))
        ax = plt.axes()
    
    if scatter:
        sns.scatterplot(x = fpr, y = tpr, ax = ax)
    sns.lineplot(x = fpr, y = tpr, ax = ax)
    sns.lineplot(x = [0, 1], y = [0, 1], color = 'green', ax = ax)
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
```

#### Se crea una función para representar la curva ROC y la distribución de probabilidades según el modelo.

```python
def RocAndProbDistrib(model, y_proba):
    
    classes = model.classes_
    
     # Plots the Probability Distributions and the ROC Curves One vs Rest
    plt.figure(figsize = (8,30))
    bins = [i/20 for i in range(20)] + [1]
    roc_auc_ovr = {}
    posRoc = 1
    posDiag = 3
    for i in range(len(classes)):
        # Gets the class
        c = classes[i]

        # Prepares an auxiliar dataframe to help with the plots
        df_aux = pd.DataFrame(X_test)
        df_aux['class'] = [1 if y == c else 0 for y in y_test]
        df_aux['prob'] = y_proba[:, i]
        df_aux = df_aux.reset_index(drop = True)
        
       

        # Plots the probability distribution for the class and the rest
        ax = plt.subplot(10, 2, i+posRoc)
        sns.histplot(x = "prob", data = df_aux, hue = 'class', color = 'b', ax = ax, bins = bins)
        ax.set_title(c)
        ax.legend([f"Class: {c}", "Rest"])
        ax.set_xlabel(f"P(x = {c})")

        # Calculates the ROC Coordinates and plots the ROC Curves
        ax_bottom = plt.subplot(10, 2, i+posDiag)
        tpr, fpr = get_all_roc_coordinates(df_aux['class'], df_aux['prob'])
        plot_roc_curve(tpr, fpr, scatter = False, ax = ax_bottom)
        ax_bottom.set_title("ROC Curve OvR")

        # Calculates the ROC AUC OvR
        roc_auc_ovr[c] = roc_auc_score(df_aux['class'], df_aux['prob'])
        
        if i%2 == 1:
            posRoc += 2
            posDiag += 2
            

    plt.tight_layout()
    
    return roc_auc_ovr
    
```

```python
def train_lr_for_each_label():
    coefficients = {}
    
    for i in range(10):
        model = Pipeline([('scaler', RobustScaler()), ('pca', PCA(15)), ('lr_Robust_for_label', LogisticRegression(C=2,penalty='l2',solver='newton-cg'))])
        model.fit(X_train, y_train==(i+1))
        coefficients[i] = model['lr_Robust_for_label'].coef_
    return coefficients



        
```

```python
def coef_by_label():
    coefficients = train_lr_for_each_label()
    i = 0
    fig, axs = plt.subplots(5, 2, figsize=(10, 30))
    fig.subplots_adjust(hspace=0.6)  # ajustar la altura de los espacios entre las filas
    for label, coef in coefficients.items():
        
        label += 1
        # Obtener valor absoluto de los coeficientes para la etiqueta x
        abs_coefficients = abs(coef)

        # Normalizar los valores obtenidos
        normalized_coefficients = abs_coefficients / abs_coefficients.sum()

        # Obtener porcentaje de los coeficientes para la etiqueta x
        percentage_coefficients = normalized_coefficients * 100

        # Crear df auxiliar
        df = pd.DataFrame({'%coef': percentage_coefficients[0], 'feature': range(len(percentage_coefficients[0]))})
        df['label'] = 'Label ' + str(label)

        # Crear gráfico de barras con Seaborn
        ax = sns.barplot(x='feature', y='%coef', hue='label', data=df, ax=axs[i//2][i%2])
        ax.set_title('Coeficientes más importantes para la etiqueta ' + str(label))
        
        i += 1

    plt.show()


coef_by_label()




```
